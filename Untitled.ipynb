```python
# Importimi i librarive
import pandas as pd
import yfinance as yf
from prophet import Prophet
import matplotlib.pyplot as plt

# Shkarkimi i t dhnave t aksioneve t Apple (AAPL)
stock_symbol = "AAPL"
data = yf.download(stock_symbol, start="2015-01-01", end="2025-01-01")

# Prgatitja e t dhnave pr Prophet
data.reset_index(inplace=True)  # Riorganizo pr t prfshir datn si kolon
data_for_prophet = pd.DataFrame({
    'ds': data['Date'].values,  # Siguro q sht 1-dimensional
    'y': data['Close'].values.flatten()  # Siguro q sht 1-dimensional
})
data_for_prophet['ds'] = pd.to_datetime(data_for_prophet['ds'])

# Inicializimi dhe trajnimi i modelit Prophet
model = Prophet()
model.fit(data_for_prophet)

# Krijimi i t ardhmes pr parashikim (p.sh., 1 vit m shum)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Vizualizimi i parashikimit
fig = model.plot(forecast)
plt.title(f"Parashikimi i mimit t Aksionit pr {stock_symbol}", fontsize=16)
plt.xlabel("Data", fontsize=12)
plt.ylabel("mimi ($)", fontsize=12)
plt.show()

# Vizualizimi i komponentve (trendi dhe sezonaliteti)
fig2 = model.plot_components(forecast)
plt.show()

# Ruajtja e parashikimeve n nj CSV
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv(f"{stock_symbol}_forecast.csv", index=False)

print(f"Parashikimet jan ruajtur n skedarin '{stock_symbol}_forecast.csv'.")
```
