```python
# Importimi i librarive
import pandas as pd
import yfinance as yf
from prophet import Prophet
import matplotlib.pyplot as plt

# Shkarkimi i të dhënave të aksioneve të Apple (AAPL)
stock_symbol = "AAPL"
data = yf.download(stock_symbol, start="2015-01-01", end="2025-01-01")

# Përgatitja e të dhënave për Prophet
data.reset_index(inplace=True)  # Riorganizo për të përfshirë datën si kolonë
data_for_prophet = pd.DataFrame({
    'ds': data['Date'].values,  # Siguro që është 1-dimensional
    'y': data['Close'].values.flatten()  # Siguro që është 1-dimensional
})
data_for_prophet['ds'] = pd.to_datetime(data_for_prophet['ds'])

# Inicializimi dhe trajnimi i modelit Prophet
model = Prophet()
model.fit(data_for_prophet)

# Krijimi i të ardhmes për parashikim (p.sh., 1 vit më shumë)
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Vizualizimi i parashikimit
fig = model.plot(forecast)
plt.title(f"Parashikimi i Çmimit të Aksionit për {stock_symbol}", fontsize=16)
plt.xlabel("Data", fontsize=12)
plt.ylabel("Çmimi ($)", fontsize=12)
plt.show()

# Vizualizimi i komponentëve (trendi dhe sezonaliteti)
fig2 = model.plot_components(forecast)
plt.show()

# Ruajtja e parashikimeve në një CSV
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv(f"{stock_symbol}_forecast.csv", index=False)

print(f"Parashikimet janë ruajtur në skedarin '{stock_symbol}_forecast.csv'.")
```
